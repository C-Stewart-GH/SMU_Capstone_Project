```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sc
import statsmodels.graphics.gofplots as sm
```

```python
#Plot Text Size Control
SMALL_SIZE = 22
MEDIUM_SIZE = 26
BIGGER_SIZE = 28

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the plot title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize

plt.rcParams['figure.figsize'] = (12, 7)
```

```python
sample_size = 1000000
standard_norm = np.random.normal(size=sample_size)
standard_norm2 = np.random.normal(loc=3,scale=1,size=sample_size)
standard_norm3 = np.random.normal(loc=3,scale=.5,size=sample_size)
```

```python
fig, ax = plt.subplots()
sns.kdeplot(standard_norm, ax=ax, fill=True, alpha=0.5)
sns.kdeplot(standard_norm2, ax=ax, fill=True, alpha=0.5)
plt.xlabel('Skill Rating')
plt.ylim(0,1)
plt.xlim(-4,7)
plt.title('Skill Probability Density')
```

```python
fig, ax = plt.subplots()
sns.kdeplot(standard_norm, ax=ax, fill=True, alpha=0.5)
sns.kdeplot(standard_norm3, ax=ax, fill=True, alpha=0.5)
plt.xlabel('Skill Rating')
plt.ylim(0,1)
plt.xlim(-4,7)
plt.title('Skill Probability Density')
```
